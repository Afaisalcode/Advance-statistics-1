#Statistics Advance Part 1

#Q1.What is a random variable in probability theory?

 - A random variable is a variable whose value is a numerical outcome of a random phenomenon.



#Q2.What are the types of random variables?

 -The main types are discrete random variables (countable values) and continuous random variables (values within a range).



#Q3.What is the difference between discrete and continuous distributions?

 - Discrete distributions describe probabilities for discrete random variables (e.g., number of coin flips)

 - Continuous distributions describe probabilities for continuous random variables (e.g., height).

 - Discrete distributions often use probability mass functions (PMFs), and continuous distributions use probability density functions (PDFs).



#Q4.What are probability distribution functions (PDF)?

 - A probability distribution function (PDF) is a function that describes the likelihood of a continuous random variable taking on a given value. The area under the PDF over an interval gives the probability that the variable falls within that interval.


#Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

 - A cumulative distribution function (CDF) gives the probability that a random variable is less than or equal to a certain value, F(x)=P(X≤x).

 - A PDF gives the probability density at a specific value (for continuous variables) or the probability of a specific value (for discrete variables).

 - The CDF is the integral (for continuous) or sum (for discrete) of the PDF up to that value.


#Q6.What is a discrete uniform distribution?

  - A discrete uniform distribution is a probability distribution where all possible values of a discrete random variable have an equal probability of occurring.


#Q7.What are the key properties of a Bernoulli distribution?

 - A Bernoulli distribution models a single trial with two possible outcomes: success (usually denoted by 1) with probability p, and failure (usually denoted by 0) with probability 1−p. Its mean is p, and its variance is p(1-p).


#Q8.What is the binomial distribution, and how is it used in probability?

 - The binomial distribution models the number of successes in a fixed number (n) of independent Bernoulli trials, each with the same probability of success (p). It's used to calculate the probability of a certain number of successes in situations with repeated independent trials.


#Q9. What is the Poisson distribution and where is it applied?

 - The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate (λ) and assuming events occur independently. It's applied in areas like queuing theory, traffic flow, and the number of defects in a product.



#Q10.What is a continuous uniform distribution?

 - A continuous uniform distribution is a probability distribution where all values within a specific interval [a,b] have an equal probability density.


#Q11.What are the characteristics of a normal distribution?

 - The normal distribution is a bell-shaped, symmetric distribution characterized by its mean (μ) and standard deviation (σ). It's unimodal, and the mean, median, and mode are equal. Approximately 68% of the data falls within one standard deviation of the mean, 95% within two, and 99.7% within three.


#Q12.What is the standard normal distribution, and why is it important?

 - The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It's important because any normal distribution can be standardized to the standard normal distribution using the Z-score, allowing for probabilities to be easily calculated using standard normal tables or software.


#Q13.What is the Central Limit Theorem (CLT), and why is it critical in statistics

 - The Central Limit Theorem (CLT) states that the distribution of the sample mean of independent, identically distributed random variables approaches a normal distribution as the sample size increases, regardless of the shape of the original population distribution.

 - It's critical because it allows us to make inferences about population means using sample means and to apply statistical methods that assume normality, even when the population is not normally distributed, provided the sample size is large enough


#Q14.How does the Central Limit Theorem relate to the normal distribution?

 - The CLT essentially establishes the normal distribution as the limiting distribution for sample means. It tells us that as we take larger and larger samples, the distribution of these sample means will become approximately normal, centered around the population mean.


#Q15.What is the application of Z statistics in hypothesis testing?

 - Z statistics are used in hypothesis testing to determine if there is a statistically significant difference between a sample mean and a population mean (when the population standard deviation is known) or between two sample means (when population standard deviations are known).

 - The Z-statistic quantifies how many standard deviations the sample mean is away from the hypothesized population mean.

#Q17.What are point estimates and interval estimates in statistics?

 - A point estimate is a single value used to estimate a population parameter (e.g., the sample mean as an estimate of the population mean). An interval estimate provides a range of values within which the population parameter is likely to lie, along with a degree of confidence (e.g., a confidence interval for the population mean).


#Q18.What is the significance of confidence intervals in statistical analysis?

 - Confidence intervals provide a range of plausible values for a population parameter, along with a level of confidence that the true parameter lies within that range. This gives a better understanding of the uncertainty associated with the estimate compared to a single point estimate.


#Q19.What is the relationship between a Z-score and a confidence interval?

 - Z-scores are used to determine the margin of error when constructing confidence intervals for population means when the population standard deviation is known. For a given confidence level (e.g., 95%), a corresponding Z-value (critical value) is found, which is then used to calculate the interval around the sample mean.


#Q20.How are Z-scores used to compare different distributions?

 - Z-scores standardize data points from different normal distributions to a common scale (the standard normal distribution). This allows for direct comparison of values across different datasets, as a Z-score indicates how unusual a data point is relative to its own distribution's mean and variability.



#Q21.What are the assumptions for applying the Central Limit Theorem?

 - The main assumptions for applying the Central Limit Theorem are:

 - Independence: The samples drawn from the population must be independent of each other.
 - Identical Distribution: The samples must be drawn from a population with a consistent probability distribution (though the shape doesn't have to be normal).
 - Finite Variance: The population from which the samples are drawn must have a finite variance.
 - Sample Size: While there's no strict rule, a sufficiently large sample size (often n≥30) is usually required for the CLT to hold reasonably well.

#Q23.How does a probability distribution relate to the expected outcome of a random variable?

 - A probability distribution describes all the possible values a random variable can take and their corresponding probabilities.

 - The expected value is a single number derived from this distribution that represents the average outcome we would expect if the random experiment were repeated many times.

 - It's a weighted average of all possible outcomes, where the weights are the probabilities of those outcomes.
